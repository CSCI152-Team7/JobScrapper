```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
import numpy as np
import re
```

```python
browser = webdriver.Firefox()
browser.get("https://www.careerbuilder.com/jobs?utf8=%E2%9C%93&keywords=Computer+Science&location=fresno")
```

```python
import random
import time
description=[]
place=[]
links=[]
title = browser.find_elements_by_css_selector("div.data-results-title")
title=title[4:12]#slciing

rand=random.randrange(3,8) #takes a while due to intentional delays
for i in range(len(title)):
    title[i].click()
    time.sleep(rand)
    try:
        link=browser.find_element_by_css_selector('#hide-fixed-top > a')
        temp = browser.find_elements_by_css_selector('div#jdp_description')
        location = browser.find_elements_by_css_selector('#jdp-data > div:nth-child(1) > div:nth-child(2) > div:nth-child(1) > div:nth-child(1) > div:nth-child(2) > span')
        for j in temp:
            j=j.get_attribute('innerHTML')
            description.append(j)
        for i in location:
            i=i.get_attribute('innerHTML')
            place.append(i)
        links.append(link)
    except NoSuchElementException:
        description.append(" ")
        place.append(" ")
        links.append(" ")
         
```

```python
jobtitle=[]
for i in title:
    jobtitle.append(i.get_attribute('innerHTML'))
```

```python
jblink=[]
newjoblist=[]
for newlist in jobtitle:
    newjoblist.append(newlist.strip())
jobtitle = jobtitle[:(len(description))]
description = description[:(len(description))]
place = place[:(len(description))]

# jobtitle = jobtitle[:(len(description))]


df=pd.DataFrame(jobtitle,columns=['jobtitle'])
df['description']=description
df['jobtitle']=newjoblist
df['place']=place
# df['link']=None
# # df['link']
df['link']=None
df.place = [re.sub('<[^>]+>','', str(x)) for x in df.place]
df.description = [re.sub('<[^>]+>','', str(x)) for x in df.description]
df.description = [re.sub('\n','', str(x)) for x in df.description]

df
# df.to_csv('test.csv')
# len(jobtitle)
```

```python
# temp=df.to_csrv()
r=df.to_json(orient='index')
with open('demo.json', 'w') as f:
    f.write(r)
```
