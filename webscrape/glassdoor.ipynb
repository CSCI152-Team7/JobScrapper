```python
title = driver1.find_elements_by_css_selector('a>div')

description=[]
place=[]
rand=random.randrange(3,8)
#takes a while due to intentional delays
for i in range(len(title)-1):
    time.sleep(rand)
    try:
        title[i].click()
        temp = driver1.find_elements_by_css_selector('div.desc')
        location = driver1.find_element_by_css_selector('div.location')
        for j in temp:
            j=j.get_attribute('innerHTML')
            description.append(j)
        i=location.get_attribute('innerHTML')
        place.append(i)
    except ElementClickInterceptedException:
        background=driver.find_element_by_css_selector('div.background-overlay')
        background.click()
    except NoSuchElementException:
        description.append(" ")
        place.append(" ")
    
```

```python
links=driver1.find_elements_by_css_selector('a.jobTitle')
title_jb=[t.get_attribute('innerHTML') for t in title]
link_jb=[l.get_attribute('href') for l in links]
link_jb=link_jb[1::2]
title_jb = title_jb[:(len(description))]
link_jb = link_jb[:(len(description))]
place = place[:(len(description))]

# len(place)
df=pd.DataFrame(title_jb,columns=['jobtitle'])
df['description']=description
df.description = [re.sub('<[^>]+>','', str(x)) for x in df.description]
df['place']=place
# link_jb.append("placeholder")
df['link']=link_jb
df
```
