```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
import numpy as np
import json
import io
import random
import unittest
import time
import re
```

```python
class Scrape:
    def __init__ (self):
        self.profile=webdriver.FirefoxProfile()
        self.profile.set_preference("permissions.default.image", 2)
        self.browser = webdriver.Firefox(self.profile)

    def crawl(self,url,name,org,desc,slry,plce):
        organization=[]
        description=[]
        place=[]
        salary=[]
        title=[]
        links=[]
        rand=random.randrange(7,12)
        
        self.browser.get(url)
        jb_name = self.browser.find_elements_by_css_selector(name)
        orgn = self.browser.find_elements_by_css_selector(org)
        for j in jb_name:
            j = j.get_attribute('innerHTML')
            title.append(j)
#             print("hey")
        for o in orgn:
            o = o.get_attribute('innerHTML')
            organization.append(o)
#             print("heyoo")

        for i in range(len(jb_name)): 
#         for i in range(4):
            time.sleep(rand)
            try: 
                jb_name[i].click()
                time.sleep(1)
                city = self.browser.find_element_by_css_selector(plce)
                
                p = city.get_attribute('innerHTML')
                place.append(p) 
                
                try:
                    descr = self.browser.find_element_by_css_selector(desc)
                    d = descr.get_attribute('innerHTML')    
                    description.append(d)
                
                except: 
                    print("ffailed dscr2")
                    description.append(" ")
                
                try:
                    sala = self.browser.find_elements_by_css_selector(slry)
                    s = sala[2].get_attribute('innerHTML')
                    salary.append(s)
                except:
                    salary.append(" ")
                    
            except:
                time.sleep(rand)
                try:
                    xbutton = self.browser.find_element_by_css_selector('div.popover-x')
                    xbutton.click()
                except:
                    pass
    
        links=[ l.get_attribute('href') for l in jb_name]
        self.organization=organization
        self.description=description
        self.place=place
        self.salary=salary
        self.title=title
        self.links=links
    
    def toDf(self):
        self.title = self.title[:(len(self.place))]
        self.links = self.links[:(len(self.place))]
        self.organization = self.organization[:(len(self.place))]
        
        df=pd.DataFrame(self.title,columns=['JobTitle'])
        df['organization']=self.organization
        df['description']=self.description
        df['salary']=self.salary
        df['place']=self.place
        df['link']=self.links
        df['site']="Indeed"
        
        df.place = [re.sub('<[^>]+>','', str(x)) for x in df.place]
        df.description = [re.sub('<[^>]+>','', str(x)) for x in df.description]
        df.organization = [re.sub('<[^>]+>','', str(x)) for x in df.organization]
        df = df.replace('\n','', regex=True)
#         df.JobTitle = df["JobTitle"].str.replace("\n", "")

        return df

def toJsonFile(df):
    x={}
    #keys JobTitle 	organization 	description 	salary 	place 	link 	site
    for i in range(len(df)):
        temp={}
        temp["model"] = "jobs.jobsinfo"
        temp["pk"]= i
        x[i]=temp
        x[i]["fields"]={
            "JobTitle":(df["JobTitle"][i]),
            "organization": df["organization"][i],
            "description":df["description"][i],
            "salary":df["salary"][i],
            "place":df["place"][i],
            "link":df["link"][i],
            "site":df["site"][i],
        }
    x=list(x.values())
    with io.open('indeed_scrape.json', 'w',encoding='windows-1252') as fp:
        json.dump(x,fp)

j=Scrape()
#self,url,name,org,desc,slry,plc
```
