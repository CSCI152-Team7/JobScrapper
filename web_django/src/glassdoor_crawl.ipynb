```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
import json
import io
import numpy as np
import random
import time
import unittest
import re
```

```python
url="https://www.indeed.com/q-computer-science-l-'Fresno'-jobs.html"
profile=webdriver.FirefoxProfile()
profile.set_preference("permissions.default.image", 2)
driver1 = webdriver.Firefox(profile)
driver1.get("https://www.glassdoor.com/Job/fresno-computer-science-jobs-SRCH_IL.0,6_IC1146705_KE7,23.htm")
```

```python
organization = driver1.find_elements_by_css_selector('a>div')
description=[]
place=[]
salary=[]
title=[]

rand=random.randrange(7,12)
#takes a while due to intentional delays
# for i in range(len(organization)-1):
for i in range(15):
    time.sleep(rand)
    try:
        organization[i].click()
        temp = driver1.find_elements_by_css_selector('div.desc')
        location = driver1.find_element_by_css_selector('div.location')
        sal = driver1.find_element_by_css_selector('div.salary')
        jobname = driver1.find_element_by_css_selector('div.title')
        
        for j in temp:
            j=j.get_attribute('innerHTML')
            description.append(j)
        i=location.get_attribute('innerHTML')
        k=sal.get_attribute('innerHTML')
        l=jobname.get_attribute('innerHTML')
        
        place.append(i)
        salary.append(k)
        title.append(l)
        
    except NoSuchElementException:
        description.append(" ")
        place.append(" ")
        salary.append(" ")
    except:
#         element = driver1.find_element_by_css_selector('.jaCreateAccountTitle')
        webdriver.ActionChains(driver1).send_keys(Keys.ESCAPE).perform()
        webdriver.ActionChains(driver1).send_keys(Keys.ESCAPE).perform()

        
links=driver1.find_elements_by_css_selector('a.jobTitle')
link_jb=[l.get_attribute('href') for l in links]
link_jb=link_jb[1::2]
```

```python
# s = [((re.search('<span class="gray salary">(.*)<span class="downsized">', x)).group(1))
#     for x in salary]
title_jb=[t.get_attribute('innerHTML') for t in organization]

title_jb = title_jb[:(len(description))]
link_jb = link_jb[:(len(description))]
place = place[:(len(description))]
title = title[:(len(description))]
# description=description[:(len(description))]

# len(place)
df=pd.DataFrame(title,columns=['JobTitle'])
df['organization']=title_jb
df['description']=description
df.description = [re.sub('<[^>]+>','', str(x)) for x in df.description]
df['salary']=salary
df.salary=[re.sub('<[^>]+>','', str(x)) for x in df.salary]
df['place']=place
# link_jb.append("placeholder")
df['link']=link_jb
# df.reset_index()
df['site']="glassdoor"
df
# len(title)
# salary[0]
```

```python
import io
x={}
#keys JobTitle 	organization 	description 	salary 	place 	link 	site
for i in range(len(df)):
    temp={}
    temp["model"] = "jobs.jobsinfo"
    temp["pk"]= i
    x[i]=temp
    x[i]["fields"]={
        "JobTitle":(df["JobTitle"][i]),
        "organization": df["organization"][i],
        "description":df["description"][i],
        "salary":df["salary"][i],
        "place":df["place"][i],
        "link":df["link"][i],
        "site":df["site"][i],
    }
x=list(x.values())
    
with io.open('newfields.json', 'w',encoding='windows-1252') as fp:
    json.dump(x,fp)
```
